###**There are null values so we fill them with mean median mode or with drop**###

```python
med=df1['Gender'].mode()[0]
df1['Gender']=df1['Gender'].fillna(med)
```

```python
df1=df1.dropna(subset=["Married"] ,axis=0)
```

```python
df1['Dependents']=df1['Dependents'].fillna(2,inplace=False)
```

```python
df1=df1.dropna(subset=["Self_Employed"] ,axis=0)
```

```python
med=df1['LoanAmount'].median()
df1['LoanAmount']=df1['LoanAmount'].fillna(med)
```

```python
med=df1['Loan_Amount_Term'].median()
df1['Loan_Amount_Term']=df1['Loan_Amount_Term'].fillna(med)
```

```python
med=df1['Credit_History'].median()
df1['Credit_History']=df1['Credit_History'].fillna(med)
```

#### **The Dependents as "+3" categorized and numeric both..** ####
#### **we can't label encode so we replace such data in Dependents column** ####

```python
df1['Dependents'] = df1['Dependents'].replace('3+', 3)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=["Loan_ID","Gender","Married","Education","Self_Employed","Property_Area","Loan_Status"]
for f in cols:
  df1[f]=le.fit_transform(df1[f])
```

# **Classification Algorithms** #

#**Logistic algorithm**

```python
x=df1[["Loan_ID","Gender","Married","Dependents","Education","Self_Employed","ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term","Credit_History","Property_Area"]]
y=df1[["Loan_Status"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
ypred=lr.predict(x_test)
print(ypred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,ypred))
```

### **We got 76% as a good accuracy** ###

##**Let's use our Logistic algorithm model**##

### we got 76 percent accuracy as a good score so we build real case  ###

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
new_applicant = np.array([[1, 1, 0, 0, 0, 0, 5000, 0.0, 150.0, 360.0, 1, 2]])
result = lr.predict(new_applicant)
if result[0] == 1:
    print("Loan is likely to be approved.")
else:
    print("Loan is likely to be rejected.")
```
